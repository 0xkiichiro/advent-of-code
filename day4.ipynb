```python
with open('day4_data.txt', 'r') as file:
    file_contents = file.read()
```

```python
raw_list = file_contents.split('\n')
for i in range(len(raw_list)):
    raw_list[i] = raw_list[i].split(',')
```

```python
counter = 0
for item in raw_list:
    first_start = int(item[0].split('-')[0])
    first_end = int(item[0].split('-')[1])
    last_start = int(item[1].split('-')[0])
    last_end = int(item[1].split('-')[1])
    print(first_start, first_end, '-', last_start, last_end)

    if (first_start <= last_start and first_end >= last_end):
        counter += 1
        print(first_start, '<= ', last_start, ' and ', first_end, '>=', last_end, counter)
    elif (last_start <= first_start and first_end <= last_end):
        counter += 1
        print(last_start, '<= ', first_start, ' and ', first_end, '<=', last_end, counter)
```

```python
print(counter)
```

```python
# part - 2

counter = 0
for item in raw_list:
    first_start = int(item[0].split('-')[0])
    first_end = int(item[0].split('-')[1])
    last_start = int(item[1].split('-')[0])
    last_end = int(item[1].split('-')[1])
    print(first_start, first_end, '-', last_start, last_end)
    if first_start <= last_end and last_start <= first_end:
        counter += 1
        print("Overlap found:", counter)
```
