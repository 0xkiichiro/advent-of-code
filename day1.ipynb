```python
# Open the file in read mode
with open('data.txt', 'r') as file:
    # Read all the contents of the file
    file_contents = file.read()

# Print the contents of the file
print(file_contents)
```

```python
split_contents = file_contents.split('\n')
```

```python
print(split_contents)
```

```python
elves = []
elf = []
for i in split_contents:
    if i == '':
        elves.append(elf)
        elf = []
        pass
    else:
        i = int(i)
        elf.append(i)
print(elves)
```

```python
elves_total = []
for elf in elves:
    elves_total.append(sum(elf))
```

```python
elves_total.sort(reverse=True)
print(elves_total)
```
