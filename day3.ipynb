```python
with open('day3_data.txt', 'r') as file:
    file_contents = file.read()
```

```python
# move items to a list
rucksacks = file_contents.split('\n')
```

```python
# split each rucksack to 2 compartments
rucksacks_splitted = []
for rucksack in rucksacks:
    print(rucksack)
    compartments = []
    length_ruck = rucksack.__len__() // 2
    print(length_ruck)
    compartments.append(rucksack[:length_ruck])
    compartments.append(rucksack[length_ruck:])
    rucksacks_splitted.append(compartments)
```

```python
# compare both lists for shared letters (case sensitive)
for rucksack in rucksacks_splitted:
    print(rucksack[0])
    print(rucksack[1])
```
