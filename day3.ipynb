```python
with open('day3_data.txt', 'r') as file:
    file_contents = file.read()
```

```python
# move items to a list
rucksacks = file_contents.split('\n')
```

```python
# split each rucksack to 2 compartments
rucksacks_splitted = []
for rucksack in rucksacks:
    print(rucksack)
    compartments = []
    length_ruck = rucksack.__len__() // 2
    compartments.append(rucksack[:length_ruck])
    compartments.append(rucksack[length_ruck:])
    rucksacks_splitted.append(compartments)
```

```python
shared_chars = []
# compare both lists for shared letters (case sensitive)
for rucksack in rucksacks_splitted:
    for char in rucksack[0]:
        if char in rucksack[1]:
            print('shared character: ', char, 'from ', rucksack[0], ' & ', rucksack[1])
            # store shared letter instead of printing
            shared_chars.append(char)
            break
```

```python
alphabet = 'abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ'
scores = []
for char in shared_chars:
    score = int(alphabet.index(char)) + 1
    scores.append(score)
```

```python
# 3 elves all share 1 common good = badge_char

# group elves in group of threes
# iterate each character from each elf to compare with rest
# store the shared letter for each group (100 letters)
# calculate prio points for each group
# sum the points
```

```python
counter = 0
grp_of_three = []
sorted_list = []
for elf in rucksacks:
    if counter == 2:
        grp_of_three.append(elf)
        sorted_list.append(grp_of_three)
        grp_of_three = []
        counter = 0
    else:
        grp_of_three.append(elf)
        counter += 1
```

```python
print(rucksacks[:10])
print('------')
print(sorted_list[:3])
print(sorted_list.__len__())
```

```python
grp_prios = []
done_list = []

for grp in sorted_list:
    for char in grp[0]:
        if (char in grp[1]) and (char in grp[2] and grp not in done_list):
            print(char, 'from ', grp)
            done_list.append(grp)
            grp_prios.append(char)
```

```python
scores_grp = []
for char in grp_prios:
    score = int(alphabet.index(char)) + 1
    scores_grp.append(score)
```
