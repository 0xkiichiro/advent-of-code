```python
# Open the file in read mode
with open('day2_data.txt', 'r') as file:
    # Read all the contents of the file
    file_contents = file.read()

# Print the contents of the file
print(file_contents)
```

```python
turns = file_contents.split('\n')
```

```python
turns_sorted = []
for i in turns:
    turns_sorted.append(i.split(' '))
turns_sorted
```

```python
# A rock, B paper, C scissors
# X rock, Y Paper, Z scissors
# 1 rock, 2 paper, 3 scissors
# 0 lost, 3 draw, 6 win

def play(opponent, me):
    points_me = 0
    # when opponent is paper
    # paper vs scissors
    if opponent == 'B' and me == 'Z':
        # me wins
        points_me += 6
        #me  scissors
        points_me += 3
    # paper vs rock
    elif opponent == 'B' and me == 'X':
        # me rock
        points_me += 1
    # paper vs paper
    elif opponent == 'B' and me == 'Y':
        # draw
        points_me += 3
        #me paper
        points_me += 2
    #when opponent is rock
    # rock vs scissors
    elif opponent == 'A' and me == 'Z':
        # me scissors
        points_me += 3
    # rock vs paper
    elif opponent == 'A' and me == 'Y':
        # me win
        points_me += 6
        # me paper
        points_me += 2
    # rock vs rock
    elif opponent == 'A' and me == 'X':
        # draw
        points_me += 3
        # me rock
        points_me += 1
    #when opponent is scissors
    # scissors vs rock
    elif opponent == 'C' and me == 'X':
        # me win
        points_me += 6
        # me rock
        points_me += 1
    # scissors vs paper
    elif opponent == 'C' and me == 'Y':
        #me paper
        points_me += 2
    # scissors vs scissors
    elif opponent == 'C' and me == 'Z':
        # draw
        points_me += 3
        # me scissors
        points_me += 3
    return points_me
```

```python
total_points = []
for opponent, you in turns_sorted:
    print(opponent, you)
    game_point = play(opponent, you)
    total_points.append(game_point)
```

```python
total_points
sum(total_points)
```

```python
# --- Part Two ---
# The Elf finishes helping with the tent and sneaks back over to you. "Anyway, the second column says how the round needs to end: X means you need to lose, Y means you need to end the round in a draw, and Z means you need to win. Good luck!"

# The total score is still calculated in the same way, but now you need to figure out what shape to choose so the round ends as indicated. The example above now goes like this:

# In the first round, your opponent will choose Rock (A), and you need the round to end in a draw (Y), so you also choose Rock. This gives you a score of 1 + 3 = 4.
# In the second round, your opponent will choose Paper (B), and you choose Rock so you lose (X) with a score of 1 + 0 = 1.
# In the third round, you will defeat your opponent's Scissors with Rock for a score of 1 + 6 = 7.
# Now that you're correctly decrypting the ultra top secret strategy guide, you would get a total score of 12.

# Following the Elf's instructions for the second column, what would your total score be if everything goes exactly according to your strategy guide?
```

```python
# A rock, B paper, C scissors
# X lose, Y draw, Z win
# 1 rock, 2 paper, 3 scissors
# 0 lost, 3 draw, 6 win

# ------- old rules below -------
# X rock, Y Paper, Z scissors

def decide_my_move(opponent, hint):
    # opponent is A = Rock
    if opponent == 'A':
        if hint == 'X':
            return 'Z'
        if hint == 'Y':
            return 'X'
        if hint == 'Z':
            return 'Y'
    elif opponent == 'B':
        if hint == 'X':
            return 'X'
        if hint == 'Y':
            return 'Y'
        if hint == 'Z':
            return 'Z'
    elif opponent == 'C':
        if hint == 'X':
            return 'Y'
        if hint == 'Y':
            return 'Z'
        if hint == 'Z':
            return 'X'
```

```python
total_points = []
for opponent, hint in turns_sorted:
    my_move = decide_my_move(opponent, hint)
    print(opponent, my_move)
    game_point = play(opponent, my_move)
    total_points.append(game_point)
```

```python
print(total_points, sum(total_points))
```
